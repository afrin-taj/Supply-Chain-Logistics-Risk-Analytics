```python
#checking for null values 
df.isnull().sum()
```

```python
#dropping 2 columns with nulll values 
df = df.drop(columns=['Product Description','Order Zipcode'])
```

```python
#connectinng notebook to sql 
from sqlalchemy import create_engine

user = "root"
password = "Mysql@123"
host = "localhost"
port = 3306
database = "supply_chain_db"

engine = create_engine(
    f"mysql+pymysql://{user}:{password.replace('@','%40')}@{host}:{port}/{database}"
)
```
